```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]
plt.bar(x,y)

#label x and y axes
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,0.5]
veh_type = ['bicycle','motorbike','car','van','stroller']

#specifying label parameter on pie chart
plt.pie(z, labels = veh_type)
plt.show()
```

```python
address = '/Users/anaygupta/Desktop/Lynda/Exercise Files/Ch01/01_05/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg #or cars['mpg']

fig = plt.figure()
#create blank figure object
ax = fig.add_axes([.1,.1,1,1])
#add axes 

mpg.plot()
ax.set_xticks(range(32))
#generate a series of tickmarks between 1 and 31 therefore range(32)

ax.set_xticklabels(cars.car_names, rotation = 60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars')

ax.set_xlabel()
```
